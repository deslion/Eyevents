```python
from eyevents.trajectories import Trajectories
```

## Загрузка всех траекторий из директории

```python
EXPERIMENT_DATA_DIR = '../data/debug/debug_files/'

settings = dict(
    columns=dict(time = 't', porx = 'x', pory = 'y'),
    common=dict(
        size=[59, 33],
        resolution=[1920, 1080],
        distance=50,
        adjust_time=True, 
        normalized=True,
        reference_point=None,
    ),
    velocity=dict(
        velocity_type='analytical',
        window=15
    ),
    loading = dict(sep=',', decimal='.', skiprows=0, header=0),
    smoothing = dict(
        window=15,
        center=False,
        method='median',
        order=2,
        fillna=True
    ),
    oculus = dict(
        method='IVT',
        velocity_threshold=30,
    )
)

trjs = Trajectories(EXPERIMENT_DATA_DIR, settings)
```
